# Minimising the Rosenbrock function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rosenbrock(X, Y):
    return ((1-X)**2)+100*(Y-X**2)**2
```

```python
xx = np.linspace(-2,2,100)
yy = np.linspace(-2,2,100)
X, Y = np.meshgrid(xx,yy)
```

```python
Z = rosenbrock(X, Y)
```

```python
%matplotlib inline
plt.figure()
levels = [0.5,1.0,2.0,5.0,10.0,50.0,100.0,200.0,600.0,800.0]
cp = plt.contourf(X,Y,Z,levels)
#plt.clabel(cp, inline=1, fontsize=10)
plt.title('Level sets of Rosenbrock function')
plt.xlabel('x')
plt.ylabel('y')
plt.plot([1],[1],'ro')
plt.show()
```
